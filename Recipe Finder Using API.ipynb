```python
class Recipe:
    def __init__(self, data):
        self.id = data.get("idMeal")
        self.name = data.get("strMeal")
        self.category = data.get("strCategory")
        self.area = data.get("strArea")
        self.instructions = data.get("strInstructions")
        self.ingredients = self._extract_ingredients(data)

    def _extract_ingredients(self, data):
        items = []
        for i in range(1, 21):
            ing = data.get(f"strIngredient{i}")
            mea = data.get(f"strMeasure{i}")
            if ing and ing.strip():
                items.append(f"{ing.strip()} - {mea}")
        return items

    def display(self):
        print("\n========== RECIPE DETAILS ==========")
        print("Name     :", self.name)
        print("Category :", self.category)
        print("Area     :", self.area)
        print("\nIngredients:")
        for ing in self.ingredients:
            print(" -", ing)
        print("\nInstructions:\n", self.instructions)
        print("====================================\n")

```

```python
def search_offline_recipes(ingredient):
    """
    Offline data se ingredient ke base par recipes search.
    Agar match na mile to sab recipes return.
    """
    meals = FAKE_API_DATA["meals"]
    ingredient = ingredient.strip().lower()

    matched = []

    for m in meals:
        r = Recipe(m)
        all_ings = " ".join(r.ingredients).lower()
        if ingredient and ingredient in all_ings:
            matched.append(r)

    if matched:
        print("Using OFFLINE data.\n")
        return matched
    else:
        if ingredient:
            print("No exact match found in OFFLINE data. Showing all dishes.\n")
        return [Recipe(m) for m in meals]
```

```python
def search_online_recipes(ingredient):
    """
    Online TheMealDB API se recipes lene ki koshish.
    Error ya no-data par empty list return karega.
    """
    try:
        # API GET with parameter ?i=ingredient
        resp = requests.get(
            f"{BASE_URL}/filter.php",
            params={"i": ingredient},
            timeout=5
        )
        resp.raise_for_status()
        data = resp.json()

        meals = data.get("meals")
        if not meals:
            return []

        recipes = []

        for meal in meals:
            meal_id = meal["idMeal"]
            detail_resp = requests.get(
                f"{BASE_URL}/lookup.php",
                params={"i": meal_id},
                timeout=5
            )
            detail_resp.raise_for_status()
            full_data = detail_resp.json()["meals"][0]

            # Optional: filter out some non-veg categories
            bad_categories = {"Beef", "Chicken", "Pork", "Seafood", "Lamb"}
            if full_data.get("strCategory") in bad_categories:
                continue

            recipes.append(Recipe(full_data))

        if recipes:
            print("Using ONLINE API data.\n")
        return recipes

    except Exception:
        # koi bhi network / API error
        return []
```

```python
def save_recipe_to_file(recipe, filename="Recipe_Output.txt"):
    try:
        with open(filename, "w", encoding="utf-8") as f:
            f.write(f"Name: {recipe.name}\n")
            f.write(f"Category: {recipe.category}\n")
            f.write(f"Area: {recipe.area}\n\n")

            f.write("Ingredients:\n")
            for ing in recipe.ingredients:
                f.write(f" - {ing}\n")

            f.write("\nInstructions:\n")
            f.write(recipe.instructions)

        print(f"Recipe saved to {filename}")
    except Exception as e:
        print("Error while saving file:", e)


def get_recipes(ingredient):
    """
    Pehle online try, fir offline fallback.
    """
    # 1) Try online
    online_list = search_online_recipes(ingredient)
    if online_list:
        return online_list

    # 2) If online fail or empty, use offline
    return search_offline_recipes(ingredient)
```

```python
def run_app():
    print("=== RECIPE FINDER USING API ===")
    ingredient = input("Enter ingredient name (e.g., potato, paneer, rice): ")

    recipes = get_recipes(ingredient)

    if not recipes:
        print("No recipes found.")
        return

    print(f"\nFound {len(recipes)} recipes:\n")
    for i, r in enumerate(recipes, start=1):
        print(f"{i}. {r.name}")

    try:
        choice = int(input("\nEnter recipe number to view details: "))
        if choice < 1 or choice > len(recipes):
            print("Invalid choice.")
            return

        selected = recipes[choice - 1]
        selected.display()

        save = input("Do you want to save this recipe to file? (y/n): ").lower()
        if save == "y":
            save_recipe_to_file(selected)
        else:
            print("Recipe not saved.")

    except ValueError:
        print("Please enter a valid number.")


run_app()
```
